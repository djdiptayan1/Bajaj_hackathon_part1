```python
import requests
import xml.etree.ElementTree as ET
import json
```

```python
response = requests.get('https://www.bajajfinserv.in/in.sitemap.en-sitemap.xml')
```

```python
root = ET.fromstring(response.content)
```

```python
namespace = {'ns': 'http://www.sitemaps.org/schemas/sitemap/0.9'}
```

```python
# Create a list to store the data
url_data_list = []
```

```python
for url in root.findall('.//ns:url', namespace):
    loc = url.find('ns:loc', namespace).text
    lastmod = url.find('ns:lastmod', namespace).text

    url_data_list.append({'URL': loc, 'last_modified': lastmod})

    print(f'URL: {loc}, Last Modified: {lastmod}')
```

```python
# Write the data to a JSON file
with open('sitemap_data.json', 'w') as json_file:
    json.dump(url_data_list, json_file, indent=2)

print('Data has been successfully stored in sitemap_data.json')
```
